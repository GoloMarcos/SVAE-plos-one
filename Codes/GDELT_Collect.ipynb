%%bigquery *NOMDE DO SEU DATAFRAME* --project *NOME DO PROJETO NO BIG QUERY DA CLOUD*

por exemplo, meu dataframe tem nome df e meu projeto no big query da cloud tem nome maiopro, meu comando ficou:

%%bigquery df --project maiopro

```
%%bigquery df --project maiopro
SELECT
  Min(GKGRECORDID) Id,
  Min(DATE(_PARTITIONTIME)) Date,
  REGEXP_EXTRACT(Extras, '<PAGE_TITLE>(.*)</PAGE_TITLE>') Title,
FROM
  `gdelt-bq.gdeltv2.gkg_partitioned`
WHERE
  EXTRACT(MONTH FROM DATE(_PARTITIONTIME)) = 05 and
  EXTRACT(YEAR FROM DATE(_PARTITIONTIME)) = 2023 and
  REGEXP_CONTAINS(Extras, '.*TITLE.*') and
  (REGEXP_CONTAINS(REGEXP_EXTRACT(Extras, '<PAGE_TITLE>(.*)</PAGE_TITLE>'), 'Agriculture ') or 
  REGEXP_CONTAINS(REGEXP_EXTRACT(Extras, '<PAGE_TITLE>(.*)</PAGE_TITLE>'), ' agriculture '))
GROUP BY Title
ORDER BY Date ASC 
```
